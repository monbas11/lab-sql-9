```python
def compare_rentals(df3):
    
    #datatime format for rental date
    df3['rental_date'] = pd.to_datetime(df3['rental_date'])

    # filer for may and june
    may_data = df3[df3['rental_date'].dt.month == 5]
    june_data = df3[df3['rental_date'].dt.month == 6]

    # count the number of rentals on may and june for each rental_id
    may_rentals = may_data['rental_id'].value_counts()
    june_rentals = june_data['rental_id'].value_counts()

    # concatenate the data
    rental_comparison = pd.concat([may_rentals, june_rentals], axis=1, keys=['rental_id_may', 'rental_id_june'])
    
    # fill NaN values with zero 
    rental_comparison.fillna(0, inplace=True)

    # calculate the difference between May and June rentals
    rental_comparison['diff'] = rental_comparison['rental_id_may'] - rental_comparison['rental_id_june']

    return rental_comparison


rental_comparison = compare_rentals(df3)

rental_comparison
```
